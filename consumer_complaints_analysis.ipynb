```python
plt.figure(figsize=(12,6))

sns.countplot(
    data=df,
    y='product',
    order=df['product'].value_counts().index,
    palette='viridis'
)

plt.title("Number of Complaints by Product")
plt.xlabel("Complaint Count")
plt.ylabel("Product")

plt.tight_layout()
plt.show()
```

```python
product_summary = pd.DataFrame({
    "Complaint Count": df['product'].value_counts(),
    "Percentage": (
        df['product']
          .value_counts(normalize=True)
          .mul(100)
          .round(2)
    )
})

product_summary
```

```python
product_counts = df['product'].value_counts()

plt.figure(figsize=(12,6))

sns.barplot(
    x=product_counts.values,
    y=product_counts.index,
    palette="viridis"
)

plt.title("Number of Customer Complaints by Product")
plt.xlabel("Complaint Count")
plt.ylabel("Product")

plt.tight_layout()
plt.show()
```

```python
product_percent = (
    df['product']
      .value_counts(normalize=True)
      .mul(100)
      .round(2)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=product_percent.values,
    y=product_percent.index,
    palette="magma"
)

plt.xlabel("Percentage of Complaints")
plt.ylabel("Product")
plt.title("Percentage Distribution of Customer Complaints")

plt.show()
```

```python
top5 = df['product'].value_counts().head(5)

plt.figure(figsize=(8,8))

plt.pie(
    top5.values,
    labels=top5.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Top 5 Products by Complaint Share")

plt.show()
```

```python
yearly = df.groupby('year').size().reset_index(name='complaints')

plt.figure(figsize=(10,5))
sns.lineplot(data=yearly, x='year', y='complaints', marker='o')

plt.title("Yearly Customer Complaint Trend")
plt.xlabel("Year")
plt.ylabel("Number of Complaints")
plt.grid(True)

plt.show()
```

```python
month_order = [
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
]

monthly = df.groupby('month').size().reindex(month_order)

plt.figure(figsize=(12,5))

sns.lineplot(
    x=monthly.index,
    y=monthly.values,
    marker='o'
)

plt.xticks(rotation=45)
plt.title("Monthly Complaint Trend")
plt.xlabel("Month")
plt.ylabel("Complaints")

plt.show()
```

```python
day_order = [
    "Monday","Tuesday","Wednesday",
    "Thursday","Friday","Saturday","Sunday"
]

daily = df.groupby('day').size().reindex(day_order)

plt.figure(figsize=(10,5))

sns.barplot(
    x=daily.index,
    y=daily.values,
    palette='crest'
)

plt.xticks(rotation=45)

plt.title("Complaints by Day of Week")

plt.show()
```

```python
daily = (
    df.groupby('date_received')
      .size()
      .reset_index(name='complaints')
)

plt.figure(figsize=(14,6))

sns.lineplot(
    data=daily,
    x='date_received',
    y='complaints'
)

plt.title("Daily Customer Complaint Timeline")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df['response_days'],
    bins=30,
    kde=True
)

plt.title("Distribution of Company Response Days")

plt.show()
```

```python
plt.figure(figsize=(10,2))

sns.boxplot(x=df['response_days'])

plt.title("Response Days Boxplot")

plt.show()
```
