```python
home = df[[
    "date",
    "home_team",
    "home_team_score",
    "away_team_score"
]].copy()

home.columns = [
    "date",
    "team",
    "goals_scored",
    "goals_conceded"
]

home["result"] = home.apply(
    lambda x: "Win" if x["goals_scored"] > x["goals_conceded"]
    else ("Draw" if x["goals_scored"] == x["goals_conceded"] else "Loss"),
    axis=1
)

away = df[[
    "date",
    "away_team",
    "away_team_score",
    "home_team_score"
]].copy()

away.columns = [
    "date",
    "team",
    "goals_scored",
    "goals_conceded"
]

away["result"] = away.apply(
    lambda x: "Win" if x["goals_scored"] > x["goals_conceded"]
    else ("Draw" if x["goals_scored"] == x["goals_conceded"] else "Loss"),
    axis=1
)

matches = pd.concat([home, away], ignore_index=True)
```

```python
matches = matches.sort_values(["team", "date"])
```

```python
def longest_streak(results):
    longest = 0
    current = 0

    for r in results:
        if r == "Win":
            current += 1
            longest = max(longest, current)
        else:
            current = 0

    return longest

winning_streak = (
    matches
    .groupby("team")["result"]
    .apply(longest_streak)
    .sort_values(ascending=False)
)

winning_streak.head(10)
```

```python
import pandas as pd

home = df[[
    "home_team",
    "home_team_fifa_rank",
    "home_team_total_fifa_points",
    "home_team_score",
    "away_team_score",
    "home_team_strength"
]].copy()

home.columns = [
    "team",
    "fifa_rank",
    "fifa_points",
    "goals_scored",
    "goals_conceded",
    "strength"
]

home["result"] = (
    df["home_team_score"] > df["away_team_score"]
).map({True: "Win", False: "Not Win"})

away = df[[
    "away_team",
    "away_team_fifa_rank",
    "away_team_total_fifa_points",
    "away_team_score",
    "home_team_score",
    "away_team_strength"
]].copy()

away.columns = [
    "team",
    "fifa_rank",
    "fifa_points",
    "goals_scored",
    "goals_conceded",
    "strength"
]

away["result"] = (
    df["away_team_score"] > df["home_team_score"]
).map({True: "Win", False: "Not Win"})

teams = pd.concat([home, away], ignore_index=True)
```

```python
comparison = teams.groupby("team").agg(
    Avg_Strength=("strength", "mean"),
    Avg_FIFA_Rank=("fifa_rank", "mean"),
    Avg_FIFA_Points=("fifa_points", "mean"),
    Avg_Goals_Scored=("goals_scored", "mean"),
    Avg_Goals_Conceded=("goals_conceded", "mean"),
    Matches=("team", "count"),
    Wins=("result", lambda x: (x == "Win").sum())
)

comparison["Win_%"] = (
    comparison["Wins"] /
    comparison["Matches"] * 100
)

comparison = comparison.round(2)

comparison.head()
```

```python
comparison.sort_values(
    "Avg_Strength",
    ascending=False
).head(15)
```

```python
comparison["Power_Score"] = (
    comparison["Avg_Strength"] * 0.40 +
    comparison["Avg_FIFA_Points"] * 0.001 * 30 +
    comparison["Win_%"] * 0.30 +
    comparison["Avg_Goals_Scored"] * 10
)

comparison.sort_values(
    "Power_Score",
    ascending=False
).head(10)
```

```python
# Select top 10 teams by Power Score
top10 = comparison.sort_values(
    "Power_Score",
    ascending=False
).head(10).copy()

# Convert Power Score into probabilities
top10["Winning Probability (%)"] = (
    top10["Power_Score"] /
    top10["Power_Score"].sum()
) * 100

top10["Winning Probability (%)"] = top10[
    "Winning Probability (%)"
].round(2)

top10[["Power_Score", "Winning Probability (%)"]]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))

plt.pie(
    top10["Winning Probability (%)"],
    labels=top10.index,
    autopct="%1.1f%%",
    startangle=140
)

plt.title("Estimated Winning Probability of Top 10 Teams\n(Based on Power Score)")

plt.axis("equal")

plt.show()
```

```python
import numpy as np

strength = top10["Power_Score"].values

teams = top10.index.tolist()
```

```python
def simulate_tournament(teams, strengths):

    teams = list(teams)
    strengths = list(strengths)

    while len(teams) > 1:

        winners = []
        winner_strength = []

        for i in range(0, len(teams), 2):

            team1 = teams[i]
            team2 = teams[i+1]

            s1 = strengths[i]
            s2 = strengths[i+1]

            p_team1 = s1 / (s1 + s2)

            winner = np.random.choice(
                [team1, team2],
                p=[p_team1, 1-p_team1]
            )

            if winner == team1:
                winners.append(team1)
                winner_strength.append(s1)
            else:
                winners.append(team2)
                winner_strength.append(s2)

        teams = winners
        strengths = winner_strength

    return teams[0]
```

```python
top8 = comparison.sort_values(
    "Power_Score",
    ascending=False
).head(8)
```

```python
results = {}

teams = top8.index.tolist()

strengths = top8["Power_Score"].tolist()

for i in range(10000):

    winner = simulate_tournament(
        teams,
        strengths
    )

    results[winner] = results.get(winner,0)+1
```

```python
probability = (
    pd.Series(results)
      .sort_values(ascending=False)
      /10000*100
)

probability
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,9))

plt.pie(
    probability,
    labels=probability.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title(
    "Simulated 2022 FIFA World Cup Winning Probability\n(10,000 Simulations)"
)

plt.show()
```
